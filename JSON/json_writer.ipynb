```python
def containsNearbyDuplicate(nums: list[int], k: int) -> bool:
    i, j = 1, k
    for num in nums:
        print(num, nums[i : i + j])
        if num in nums[i: i + j]:
            return True
        i += 1
        # j += 1
    return False
containsNearbyDuplicate(nums = [1,0,1,1], k = 1)
```
