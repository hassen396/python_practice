```python
from animal import Animal
class Dog(Animal):
    def __init__(self, speed):
        super().__init__()
        self.speed = speed

    @staticmethod
    def pee():
        print('the dog is ***'.capitalize())
```

```python
dog = Dog(22)
dog.set_age(1)
print(f"the animal is {dog.speed} km/h fast and  {dog.age} {'years' if dog.age >=2 else 'year'} old")
dog.walk()
dog.pee()
```
