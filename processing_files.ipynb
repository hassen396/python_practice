```python
def count_words(content):
    words = content.split()
    print(len(words))
```

```python
from pathlib import Path
file_names = ['./text_files/text_file_to_read', './text_files/written.txt', './text_files/written_file.txt']
for file_name in file_names:
    try:
        content = Path(file_name).read_text()
    except :
        continue
    else:
        count_words(content)

```
