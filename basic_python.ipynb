```python
from typing import Optional

# 1. You must define the ListNode class first
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        curr = dummy
        carry = 0
        
        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0
            
            total = val1 + val2 + carry
            carry = total // 10
            
            curr.next = ListNode(total % 10)
            curr = curr.next
            
            if l1: l1 = l1.next
            if l2: l2 = l2.next
            
        return dummy.next

# --- HELPER FUNCTIONS TO RUN THE CODE ---

def to_linked_list(iterable):
    """Converts a Python list [2, 4, 3] into a ListNode chain."""
    dummy = ListNode(0)
    curr = dummy
    for val in iterable:
        curr.next = ListNode(val)
        curr = curr.next
    return dummy.next

def to_python_list(node):
    """Converts a ListNode chain back into a Python list [7, 0, 8] for printing."""
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

# --- EXECUTION ---

# Create two numbers as linked lists: 342 (2->4->3) and 465 (5->6->4)
l1 = to_linked_list([2, 4, 8])
l2 = to_linked_list([5, 6, 4])

# Initialize the solution and run the method
sol = Solution()
result_node = sol.addTwoNumbers(l1, l2)

# Convert the result back to a list to see it
print(f"Result: {to_python_list(result_node)}") 
# Output: [7, 0, 8]  (which represents 807)
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
# class Solution:
#     def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
#         output = None
#         cary = 0
#         tail= None
#         while l1 != None or l2 != None:
#             if l1 == None:
#                 sum = l2.val + cary
#             elif l2 == None:
#                 sum = l1.val + cary
#             else:
#                 sum = l1.val + l2.val + cary
#
#             if output == None:
#                 output=ListNode(sum % 10)
#             else:
#                 output.next = ListNode(sum % 10)
#                 print(output)
#             tail= output.next
#
#             cary = sum // 10
#
#             if l2 == None:
#                l1 = l1.next
#             elif l1 == None:
#                 l2 = l2.next
#             else:
#                 l1,l2 = l1.next, l2.next
#
#         print(output)
#
#         if cary:
#             output.next = ListNode(cary)
#         return output
```
